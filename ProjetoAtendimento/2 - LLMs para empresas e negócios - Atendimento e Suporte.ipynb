### 2 - Divisão em pedaços de texto / Split





```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Divide os documentos em chunks de até 500 caracteres, com 50 de overlap entre eles
# (overlap evita perder contexto que ficaria cortado na fronteira entre dois chunks).
# Esses chunks serão depois transformados em embeddings e indexados pro retriever do RAG.
# RecursiveCharacterTextSplitter tenta cortar o texto do jeito "menos ruim" possível:
# primeiro tenta separar por parágrafo (\n\n), se o pedaço ainda for maior que chunk_size
# tenta por linha (\n), depois por frase/espaço, e só em último caso corta no meio da palavra.
# chunk_size=500 -> tamanho máximo de cada pedaço (em caracteres)
# chunk_overlap=50 -> repete os últimos 50 caracteres de um chunk no início do próximo,
# pra não perder contexto que ficaria cortado bem na fronteira entre dois chunks

text_splitter = RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap=50)
chunks = []
for doc in loaded_documents:
  for doc in loaded_documents:
  # split_text(doc) retorna uma LISTA de strings (os pedaços daquele documento)
  # .extend() pega essa lista e adiciona cada item individualmente em "chunks"
  # (diferente de .append(), que adicionaria a lista inteira como um único elemento,
  # criando uma lista de listas em vez de uma lista "achatada" de strings)
  chunks.extend(text_splitter.split_text(doc))

# resultado final: "chunks" é uma lista simples (flat) de strings,
# onde cada string é um pedaço de texto de até 500 caracteres,
# pronta pra virar embedding e ser indexada no vector store do RAG
print(f"Número de chunks criados: {len(chunks)}")
```



## Etapas de Recuperação e geração

### 4 - Configurando o recuperador de texto / Retriever



```python
# Transforma o vectorstore (FAISS) em um "retriever" — objeto que recebe uma pergunta
# e devolve os chunks mais relevantes pra ela.
# search_type="similarity" -> busca pelos chunks com embedding mais próximo (similaridade
# de cosseno/distância) do embedding da pergunta, sem aplicar filtros extras de diversidade
# search_kwargs -> parâmetros extras passados pro método de busca escolhido em search_type.
# Aqui, "k": 6 define quantos chunks serão retornados: os 6 mais similares à pergunta
# (esses 6 chunks serão concatenados e viram o {context} do prompt_rag

retriever = vectorstore.as_retriever(search_type = "similarity", search_kwargs={"k": 6})
retriever
```

### 5 - Geração



```python
prompt_rag = PromptTemplate(
    input_variables=["context", "input"],
    template=template_rag,
)
prompt_rag
```

```python
# RunnablePassthrough: Runnable "neutro" que apenas devolve o input recebido sem
# nenhuma transformação. Usado quando você precisa que o input original sobreviva
# intacto até uma etapa seguinte da chain, em paralelo a outras transformações.
from langchain_core.runnables import RunnablePassthrough

# LCEL (LangChain Expression Language): o operador "|" encadeia Runnables, onde a
# saída de um vira a entrada do próximo, formando um RunnableSequence
chain_rag = (
    # Dicionário Python aqui é convertido automaticamente em um RunnableParallel.
    # Teoria do "fan-out": o MESMO input que a chain recebe (a pergunta) é replicado
    # e enviado simultaneamente para cada valor do dicionário, de forma independente.
    # No final, os resultados são recombinados num novo dicionário com as mesmas chaves.
    {
        "context": retriever,           # retriever é um Runnable que recebe a pergunta,
                                         # busca os chunks mais similares no FAISS (vector
                                         # store) e devolve uma lista de Documents

        "input": RunnablePassthrough()  # repassa a pergunta original sem alteração, pra
                                         # ela sobreviver e alimentar o {input} do prompt
    }
    # saída deste passo: {"context": [...chunks...], "input": "pergunta original"}
    # necessário pois o prompt precisa de DOIS valores (contexto + pergunta), mas a
    # chain só recebe UM input no invoke() — o RunnableParallel resolve esse "1 pra N"

    | prompt_rag        # PromptTemplate/ChatPromptTemplate: recebe o dicionário e
                        # preenche os placeholders {context} e {input} do template,
                        # gerando o prompt final (texto ou lista de mensagens)

    | llm               # recebe o prompt já formatado e gera a resposta do modelo
                        # (saída normalmente é um objeto tipo AIMessage)

    | StrOutputParser() # extrai apenas o conteúdo textual da resposta do LLM,
                        # devolvendo uma string simples (em vez do objeto de mensagem)
)
```

```python
res = chain_rag.invoke("Qual é o público alvo do banco?")
show_res(res)
```

```python
# Quando o input "Como alterar minha senha?" entra na chain, o primeiro componente é
# este dicionário, que o LangChain converte automaticamente em um RunnableParallel.
# O comportamento de um RunnableParallel é: o MESMO input recebido pela chain é
# distribuído (em paralelo) para CADA valor do dicionário — não é uma sequência,
# é uma bifurcação.
chain_rag = (
    {
        "context": retriever,            # retriever recebe "Como alterar minha senha?",
                                          # busca os k chunks mais similares no FAISS e
                                          # devolve eles como contexto (lista de Documents)

        "input": RunnablePassthrough()   # RunnablePassthrough recebe o MESMO input original
                                          # e não faz nada com ele, só repassa adiante sem
                                          # nenhuma alteração
    }
    # resultado desse primeiro passo: um novo dicionário
    # {"context": [chunk1, chunk2, ...], "input": "Como alterar minha senha?"}
    # as duas saídas (retriever + passthrough) se juntam de novo nesse dicionário
    # antes de seguir pro próximo passo da chain

    | qa_prompt         # pega esse dicionário e formata o template de prompt,
                        # substituindo {context} e {input} pelos valores correspondentes
    | llm               # envia o prompt já formatado pro modelo, que gera a resposta
                        # com base no contexto recuperado
    | StrOutputParser() # extrai só o texto da resposta (string simples)
)

res = chain_rag.invoke("Como alterar minha senha?")
show_res(res)
```

## Pipeline RAG avançada



### 1. Reformulação da consulta para contextualizar

Em resumo teremos duas chains, uma para contextualização e outra para resposta

```python
# MessagesPlaceholder: permite injetar uma LISTA de mensagens (o histórico da conversa)
# dentro de um ChatPromptTemplate, em vez de só um texto fixo num placeholder comum
from langchain_core.prompts import MessagesPlaceholder

# AIMessage / HumanMessage: classes que representam mensagens trocadas na conversa,
# usadas pra montar o "chat_history" (ex: HumanMessage("oi") , AIMessage("olá!"))
from langchain_core.messages import AIMessage, HumanMessage

# create_history_aware_retriever: helper do LangChain que cria um retriever "inteligente",
# capaz de levar em conta o histórico da conversa antes de buscar no vector store
# create_retrieval_chain: helper que monta a chain RAG completa (retrieval + geração)
from langchain.chains import create_history_aware_retriever, create_retrieval_chain

# create_stuff_documents_chain: helper que cria a etapa de "juntar os documentos
# recuperados e gerar a resposta final" (o "stuff" = enfiar todos os chunks recuperados
# direto no prompt, sem sumarização ou estratégia mais complexa)
from langchain_classic.chains.combine_documents import create_stuff_documents_chain

# Esse prompt NÃO é o prompt que responde a pergunta do usuário. É um prompt auxiliar,
# usado só pra REFORMULAR a pergunta levando em conta o histórico da conversa.
# Resolve o problema clássico do RAG conversacional: se o usuário pergunta "e essa segunda?"
# depois de já ter perguntado sobre algo, o retriever sozinho não entende a que "essa" se
# refere -- ele só recebe a última pergunta isolada, sem contexto da conversa.
context_q_system_prompt = "Given the following chat history and the follow-up question which might reference context in the chat history, formulate a standalone question which can be understood without the chat history. Do NOT answer the question, just reformulate it if needed and otherwise return it as is."

context_q_system_prompt = context_q_system_prompt  # linha redundante, não faz nada (reatribuição a si mesma)

context_q_user_prompt = "Question: {input}"  # placeholder {input} será preenchido com a pergunta atual do usuário

# Monta o template de chat com 3 partes, na ordem em que o LLM vai "ler":
context_q_prompt = ChatPromptTemplate.from_messages(
    [
        ("system", context_q_system_prompt),   # instrução: reformule a pergunta sem o histórico
        MessagesPlaceholder("chat_history"),     # aqui entra a lista de mensagens anteriores
                                                  # (HumanMessage/AIMessage) trocada na conversa
        ("human", context_q_user_prompt),        # a pergunta atual (possivelmente ambígua/incompleta)
    ]
)

# Resultado: quando esse prompt for enviado a um LLM, ele devolve uma pergunta "standalone"
# (independente, sem precisar do histórico) -- essa pergunta reformulada é o que de fato
# será usada pra buscar no retriever (FAISS), em vez da pergunta crua do usuário.
# Esse é o papel do create_history_aware_retriever, que provavelmente usa esse prompt
# em seguida pra criar um retriever que "entende" o contexto da conversa.
```

```python
history_aware_retriever = create_history_aware_retriever(
    llm = llm, retriever = retriever, prompt = context_q_prompt
)
```

## Configuração a função de Indexação e Recuperação


```python
from pathlib import Path
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

def config_retriever(folder_path="/content"):
    # Carregar documentos
    docs_path = Path("/content")
    pdf_files = [f for f in docs_path.glob("*.pdf")]

    loaded_documents = [extract_text_pdf(pdf) for pdf in pdf_files]

    # Divisão em pedaços de texto / Split
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,
        chunk_overlap=200
    )
    chunks = []
    for doc in loaded_documents:
        chunks.extend(text_splitter.split_text(doc))

    # Embeddings
    embedding_model = "BAAI/bge-m3"

    embeddings = HuggingFaceEmbeddings(model_name=embedding_model)

    # Armazenamento
    vectorstore = FAISS.from_texts(chunks, embedding=embeddings)

    vectorstore.save_local('index_faiss')

    # Configurando o recuperador de texto / Retriever
    retriever = vectorstore.as_retriever(
        search_type='mmr',
        search_kwargs={'k':3, 'fetch_k':4}
    )

    return retriever
```

```python
retriever = config_retriever()
```

```python
rag_chain = config_rag_chain(llm, retriever)
```

## Junção da pipeline final + Conclusão

```python
def chat_llm(rag_chain, input, chat_history):
  chat_history.append(HumanMessage(content = input))

  response = rag_chain.invoke({
      "input": input,
      "chat_history": chat_history
  })

  res = response["answer"]

  chat_history.append(AIMessage(content = res))

  return chat_history
```

```python
%%writefile app02.py
import streamlit as st
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from pathlib import Path
from langchain_community.document_loaders import PyMuPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_core.prompts import MessagesPlaceholder
from langchain_core.messages import AIMessage, HumanMessage
from langchain.chains import create_history_aware_retriever, create_retrieval_chain
from langchain_classic.chains.combine_documents import create_stuff_documents_chain
from langchain.vectorstores import FAISS
from dotenv import load_dotenv
load_dotenv()

st.set_page_config(page_title="Atendimento SafeBank 🤖", page_icon="🤖")
st.title("Atendimento SafeBank")

id_model = "llama-3.3-70b-versatile"
temperature = 0.7
path = "/content"

### Carregamento da LLM
def load_llm(id_model, temperature):
  llm = ChatGroq(
    model=id_model,
    temperature=temperature,
    max_tokens=None,
    timeout=None,
    max_retries=2,
  )
  return llm

llm = load_llm(id_model, temperature)

### Exibição do resultado
def show_res(res):
  from IPython.display import Markdown
  if "</think>" in res:
    res = res.split("</think>")[-1].strip()
  else:
    res = res.strip()  # fallback se não houver tag
  display(Markdown(res))

### Extração do conteúdo
def extract_text_pdf(file_path):
  loader = PyMuPDFLoader(file_path)
  doc = loader.load()
  content = "\n".join([page.page_content for page in doc])
  return content

### Indexação e recuperação
def config_retriever(folder_path="/content"):
  # Carregar documentos
  docs_path = Path("/content")
  pdf_files = [f for f in docs_path.glob("*.pdf")]

  if len(pdf_files) < 1:
    st.error("Nenhum arquivo PDF carregado")
    st.stop()

  loaded_documents = [extract_text_pdf(pdf) for pdf in pdf_files]

  # Divisão em pedaços de texto / Split
  text_splitter = RecursiveCharacterTextSplitter(
      chunk_size=1000,
      chunk_overlap=200
  )
  chunks = []
  for doc in loaded_documents:
      chunks.extend(text_splitter.split_text(doc))

  # Embeddings
  embedding_model = "BAAI/bge-m3" #sentence-transformers/all-mpnet-base-v2

  embeddings = HuggingFaceEmbeddings(model_name=embedding_model)

  # Armazenamento
  vectorstore = FAISS.from_texts(chunks, embedding=embeddings)

  vectorstore.save_local('index_faiss')

  # Configurando o recuperador de texto / Retriever
  retriever = vectorstore.as_retriever(
      search_type='mmr',
      search_kwargs={'k':3, 'fetch_k':4}
  )

  return retriever

### Chain da RAG
def config_rag_chain(llm, retriever):

  # Prompt de contextualização
  context_q_system_prompt = "Given the following chat history and the follow-up question which might reference context in the chat history, formulate a standalone question which can be understood without the chat history. Do NOT answer the question, just reformulate it if needed and otherwise return it as is."

  context_q_system_prompt = context_q_system_prompt
  context_q_user_prompt = "Question: {input}"
  context_q_prompt = ChatPromptTemplate.from_messages(
      [
          ("system", context_q_system_prompt),
          MessagesPlaceholder("chat_history"),
          ("human", context_q_user_prompt),
      ]
  )

  # Chain para contextualização
  history_aware_retriever = create_history_aware_retriever(
    llm=llm, retriever=retriever, prompt=context_q_prompt
  )

  # Prompt para perguntas e respostas (Q&A)
  system_prompt = """Você é um assistente virtual prestativo e está respondendo perguntas gerais sobre os serviços de uma empresa.
  Use os seguintes pedaços de contexto recuperado para responder à pergunta.
  Se você não sabe a resposta, apenas comente que não sabe dizer com certeza.
  Mas caso seja uma dúvida muito comum, pode sugerir como alternativa uma solução possível.
  Mantenha a resposta concisa.
  Responda em português. \n\n"""

  qa_prompt = ChatPromptTemplate.from_messages([
    ("system", system_prompt),
    MessagesPlaceholder("chat_history"),
    ("human", "Pergunta: {input}\n\n Contexto: {context}"),
  ])

  # Configurar LLM e Chain para perguntas e respostas (Q&A)

  qa_chain = create_stuff_documents_chain(llm, qa_prompt)

  rag_chain = create_retrieval_chain(
    history_aware_retriever,
    qa_chain,
  )

  return rag_chain

### Interação com chat
def chat_llm(rag_chain, input):

  st.session_state.chat_history.append(HumanMessage(content=input))

  response = rag_chain.invoke({
      "input": input,
      "chat_history": st.session_state.chat_history
  })

  res = response["answer"]
  res = res.split("</think>")[-1].strip() if "</think>" in res else res.strip()

  st.session_state.chat_history.append(AIMessage(content=res))

  return res

input = st.chat_input("Digite sua mensagem aqui...")

if "chat_history" not in st.session_state:
  st.session_state.chat_history = [
      AIMessage(content = "Olá, sou o seu assistente virtual! Como posso te ajudar?"),
  ]

if "retriever" not in st.session_state:
  st.session_state.retriever = None

for message in st.session_state.chat_history:
  if isinstance(message, AIMessage):
    with st.chat_message("AI"):
      st.write(message.content)
  elif isinstance(message, HumanMessage):
    with st.chat_message("Human"):
      st.write(message.content)

if input is not None:
  with st.chat_message("Human"):
    st.markdown(input)

  with st.chat_message("AI"):
    if st.session_state.retriever is None:
      st.session_state.retriever = config_retriever(path)
    rag_chain = config_rag_chain(llm, st.session_state.retriever)
    res = chat_llm(rag_chain, input)
    st.write(res)
```
